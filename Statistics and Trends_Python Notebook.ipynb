## Data Visualisations

### Bar chart showing distribution of average Work-life balance across Industry across Work_Location

```python
plt.figure(figsize = (10,6))
sns.barplot(data = data, x = 'Industry', y = 'Work_Life_Balance_Rating', hue = 'Work_Location', errwidth = 0)
plt.title ("Distribution of average Work-life balance across Industry across Work_Location")
plt.ylabel('Work-life balance ratings')
plt.xlabel('Industry')
plt.legend(loc='upper left')
plt.show()
```

### Histogram plot showing distribution of Hours worked per week

```python
# Calculating the descriptive statistics for 'Hours_Worked_Per_Week'
stats = data['Hours_Worked_Per_Week'].describe()

# Creating the histogram with KDE
plt.figure(figsize=(14, 6))
sns.displot(data=data, x='Hours_Worked_Per_Week', binwidth=5, element='bars', fill=True, kde=True, color='lightblue')

# Adding the title and axis labels
plt.title("Distribution of Hours Worked Per Week")
plt.ylabel("Frequency")
plt.xlabel("Working Hours per Week")

# Displaying descriptive statistics on the plot
textstr = (f"Mean: {stats['mean']:.2f}\n"
           f"Median: {data['Hours_Worked_Per_Week'].median():.2f}\n"
           f"Std Dev: {stats['std']:.2f}\n"
           f"Min: {stats['min']:.2f}\n"
           f"Max: {stats['max']:.2f}")

# Placing the statistics in the top-right corner of the plot
plt.gca().text(0.95, 0.95, textstr, transform=plt.gca().transAxes,
               fontsize=10, verticalalignment='top', horizontalalignment='right',
               bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

### Scatter plot between Age and Work_Life_Balance_Rating (Segmented by Mental_Health_Condition)

```python
plt.figure(figsize = (8,6))
sns.relplot(data = data, x = 'Age', y = 'Work_Life_Balance_Rating', hue = 'Mental_Health_Condition')
plt.title("Relationship between age and Work life balance")
plt.ylabel("Work Life Balance Ratings")
plt.xlabel ("Age (Years)")
plt.show()
```

### Correlation Heatmap

```python
columns = ['Age', 'Years_of_Experience', 'Hours_Worked_Per_Week', 'Number_of_Virtual_Meetings', 'Work_Life_Balance_Rating', 'Social_Isolation_Rating']

plt.figure(figsize = (8,6))
sns.heatmap(data = data [columns].corr(), annot = True, cmap = 'Blues', cbar = True)
plt.title("Correlation Heatmap")
plt.show()
```

### Box plot

```python
# Plotting the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[columns])
plt.title("Box Plot for Numerical Features")
plt.ylabel('Distribution')
plt.xticks(rotation=45)
plt.show()
```
